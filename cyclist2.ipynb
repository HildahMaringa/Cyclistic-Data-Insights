```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os  # Import the os module

# Define the path to the datasets
folder_path = r"C:\Users\ADMIN\Desktop\Cyclist_data\Excel_files"

all_files = os.listdir(folder_path)  # Corrected function name to listdir

data_frames = []

for filename in all_files:
    file_path = os.path.join(folder_path, filename)  # Create the full path for each file
    df = pd.read_csv(file_path)  # Read each CSV file
    data_frames.append(df)  # Add the DataFrame to the list

all_data = pd.concat(data_frames, ignore_index=True)  # Concatenate all DataFrames

```

```python
for col in ['rideable_type', 'rideable_type' , 'started_at' , 'ended_at' ,'start_station_name', 'start_station_id', 
            'end_station_name', 'end_station_id', 'member_casual']:
    all_data[col] = all_data[col].astype('category')

print(all_data.info())

```

```python
# Check for missing values
missing_data = all_data.isnull().sum()
print(missing_data)

# let us how many rows are affected by missing data in the started_at, ended_at, start_station_name, start_station_id
columns_of_interest = ['started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id']
missing_counts = df[columns_of_interest].isnull().sum()
rows_with_missing_data = df[columns_of_interest].isnull().sum(axis=1).sum()
print("\n\n")
print(f"Missing values in each column:\n{missing_counts}")
print(f"\nTotal rows with missing data in any of the specified columns: {rows_with_missing_data}")
print("\n\n")


# let us analyze the pattern of missing data

# Check for missing data by year
df['started_at'] = pd.to_datetime(df['started_at'], errors='coerce')
df['year'] = df['started_at'].dt.year
missing_by_year = df.groupby('year')[['start_station_name', 'start_station_id', 'end_station_name', 'end_station_id']].apply(lambda x: x.isnull().sum()).reset_index()
print(missing_by_year)
print("\n\n")

# Check missing data by ride type (member_casual)
missing_by_ride_type = df.groupby('member_casual')[['start_station_name', 'start_station_id', 'end_station_name', 'end_station_id']].apply(lambda x: x.isnull().sum())
print(missing_by_ride_type)
print("\n\n")

# Check missing data by start and end latitude and longitude
missing_by_location = df[['start_lat', 'start_lng', 'end_lat', 'end_lng']].isnull().sum()
print(missing_by_location)

```

```python
# Convert columns to category dtype if not already
for col in ['start_station_name', 'start_station_id', 'end_station_name', 'end_station_id']:
    if all_data[col].dtype != 'category':  # Check if the column is not already category
        all_data[col] = all_data[col].astype('category')  # Convert to category dtype
    
    # Check if 'Unknown' is already a category, if not, add it
    if 'Unknown' not in all_data[col].cat.categories:
        all_data[col] = all_data[col].cat.add_categories('Unknown')

    # Replace missing values with 'Unknown'
    all_data[col] = all_data[col].fillna('Unknown')

# For the latitude and longitude columns (if they are still float), you can fill them with a placeholder string "Unknown"
# First, convert them to category and add "Unknown" as a category
for col in ['start_lat', 'start_lng', 'end_lat', 'end_lng']:
    all_data[col] = all_data[col].astype('category')
    if 'Unknown' not in all_data[col].cat.categories:
        all_data[col] = all_data[col].cat.add_categories('Unknown')  # Add 'Unknown' as a valid category
    all_data[col] = all_data[col].fillna('Unknown')  # Replace missing values with 'Unknown'

# Verify the changes
print(all_data.isnull().sum())  # Check if there are any remaining missing values
print(all_data['start_station_name'].unique())  # Check unique values for one of the columns
```

```python
# EDA

#General summary statistics
summary_stats = all_data.describe(include = "all")
print(summary_stats)
display(summary_stats)
```

```python
# Count the number of rides by each member type (Casual vs Member)
member_ride_count = all_data['member_casual'].value_counts()
print(member_ride_count)


member_ride_count.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Total Rides by Member Type')
plt.xlabel('Member Type')
plt.ylabel('Number of Rides')
plt.xticks(rotation=0)
plt.show()
```

```python
# Count the number of rides by each user (ride_id)
rides_per_user = all_data.groupby('ride_id')['member_casual'].first().value_counts()

# Calculate the average number of rides per user by member type
rides_per_user_by_type = all_data.groupby('member_casual')['ride_id'].nunique()
print(rides_per_user_by_type)

# Visualize the frequency of rides per user
rides_per_user_by_type.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Average Number of Rides per User by Member Type')
plt.xlabel('Member Type')
plt.ylabel('Average Number of Rides per User')
plt.xticks(rotation=0)
plt.show()
```

```python
# Count the number of rides by rideable type and member type
rideable_type_count = all_data.groupby(['member_casual', 'rideable_type']).size().unstack().fillna(0)
print(rideable_type_count)

# Visualize the preferred rideable types by member type
rideable_type_count.plot(kind='bar', stacked=True, color=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Preferred Rideable Types by Member Type')
plt.xlabel('Member Type')
plt.ylabel('Number of Rides')
plt.xticks(rotation=0)
plt.legend(title='Rideable Type')
plt.show()
```

```python
#  Ride Duration Analysis
#  Average Ride Duration by Member Type
# Convert 'started_at' and 'ended_at' to datetime if not already done
df['started_at'] = pd.to_datetime(df['started_at'], errors='coerce')
df['ended_at'] = pd.to_datetime(df['ended_at'], errors='coerce')

# Calculate the ride duration in minutes
df['ride_duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

# Group by member type and calculate the average ride duration
avg_ride_duration = df.groupby('member_casual')['ride_duration'].mean()

print(avg_ride_duration)

# Set up the plot
plt.figure(figsize=(10, 6))

# Plot the distribution of ride durations by member type
sns.histplot(df[df['member_casual'] == 'casual']['ride_duration'], color='blue', label='Casual', kde=True, bins=50)
sns.histplot(df[df['member_casual'] == 'member']['ride_duration'], color='green', label='Member', kde=True, bins=50)

# Customize the plot
plt.title('Distribution of Ride Durations by Member Type')
plt.xlabel('Ride Duration (minutes)')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()

```
