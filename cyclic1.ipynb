```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os  # Import the os module

# Define the path to the datasets
folder_path = r"C:\Users\ADMIN\Desktop\Cyclist_data"

all_files = os.listdir(folder_path)  # Corrected function name to listdir

data_frames = []

for filename in all_files:
    file_path = os.path.join(folder_path, filename)  # Create the full path for each file
    df = pd.read_csv(file_path)  # Read each CSV file
    data_frames.append(df)  # Add the DataFrame to the list

all_data = pd.concat(data_frames, ignore_index=True)  # Concatenate all DataFrames

```

```python
for col in ['rideable_type', 'rideable_type' , 'started_at' , 'ended_at' ,'start_station_name', 'start_station_id', 
            'end_station_name', 'end_station_id', 'member_casual']:
    all_data[col] = all_data[col].astype('category')

print(all_data.info())

```
